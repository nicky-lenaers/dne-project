# Setup Environment

```python
import gymnasium as gym
import tianshou as ts

def get_env(render_mode=None):
  return gym.make("CartPole-v1", render_mode=render_mode)

env = get_env()
```

# Create Policy

```python
import random
from tianshou.policy import BasePolicy
from tianshou.data import Batch


class RandomPolicy(BasePolicy):
  def forward(self, batch, state):
    return Batch(act=[random.randint(0, 1)])
  
  def learn():
    pass
  
policy = RandomPolicy()
```

# Play

Play using a random agent.

```python
policy.eval()

env = get_env(render_mode=None)
env = ts.env.DummyVectorEnv([lambda: env])
collector = ts.data.Collector(policy, env, exploration_noise=True)
result = collector.collect(n_episode=100, render=None)
rews, lens = result["rews"], result["lens"]

display(rews.mean())
```

Plot results

```python
import plotly.figure_factory as ff

fig = ff.create_distplot([result['rews']], ['reward'])
fig.update_layout(title_text='CartPole Random Agent Result')
fig.show()
```
